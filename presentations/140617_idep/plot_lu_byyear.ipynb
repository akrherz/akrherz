```python
import psycopg2
PGCONN = psycopg2.connect(database='idep', host='iemdb')
cursor = PGCONN.cursor()

data = {}
for i in range(1,7):
    yr = 2007 + i
    data[yr] = {}
    lu = 'landuse%s' % (i,)
    cursor.execute("""SELECT %s, count(*) from flowpath_points 
    GROUP by %s """ % (lu, lu))
    for row in cursor:
        data[yr][ row[0] ] = row[1]
```

```python
obs = {}
for line in open('nass_yearly.csv'):
    tokens = line.split(",")
    yr = int(tokens[0])
    if not obs.has_key(yr):
        obs[yr] = {}
    obs[yr][tokens[1]] = float(tokens[2])
```

```python
nass_c = []
nass_b = []
idep_c = []
idep_b = []
for yr in range(2008,2014):
    nass_c.append( obs[yr]['C'] )
    nass_b.append( obs[yr]['B'] )
    idep_c.append( data[yr]['C'] )
    idep_b.append( data[yr]['B'] )

idep_b = np.array(idep_b, 'f')
idep_c = np.array(idep_c, 'f')
nass_b = np.array(nass_b, 'f')
nass_c = np.array(nass_c, 'f')
```

```python
WEPPDB = psycopg2.connect(database='wepp', host='mesonet.agron.iastate.edu', user='nobody')
cursor = WEPPDB.cursor()
import re

cursor.execute("""
select n.id, m.name from nri n JOIN managements m on (m.man_id = n.man_id)
""")
c = [0,0,0,0]
b = [0,0,0,0]
for row in cursor:
    rotation = re.findall("[A-Z]{4}", row[1])[0]
    for p in range(4):
        if rotation[p] == 'C':
            c[p] += 1
        if rotation[p] == 'B':
            b[p] += 1

wepp_c = []
wepp_b = []
for year in range(2008,2014):
    wepp_c.append( c[(year-1997)%4] )
    wepp_b.append( b[(year-1997)%4] )
    
wepp_c = np.array(wepp_c, 'f')
wepp_b = np.array(wepp_b, 'f')
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)

ax.bar(np.arange(2008,2014)-0.4, nass_c, width=0.4, label='Corn', fc='g' )
ax.bar(np.arange(2008,2014), nass_b, width=0.4, label='Soybeans', fc='r')

y2 = ax.twinx()
y2.plot(np.arange(2008,2014), idep_c)
y2.plot(np.arange(2008,2014), idep_b)
```

```python
(fig, ax) = plt.subplots(1,1)

ax.plot(range(2008,2014), wepp_c / (wepp_c + wepp_b) * 100., lw=3, label='IDEP NRI')
ax.plot(range(2008,2014), idep_c / (idep_c + idep_b) * 100., lw=3, label='IDEPv2 Flowpaths')
ax.plot(range(2008,2014), nass_c / (nass_c + nass_b) * 100., lw=3, label='NASS Planted Acres')
ax.legend(loc=2)
ax.set_ylabel("Expressed as Percentage [%]")
ax.set_title("2008-2013 Iowa Ratio of Corn to Corn+Soybean")
ax.set_ylim(54,64)
ax.grid(True)
ax.set_xticklabels(("2008", "2009", "2010", "2011", "2012", "2013"))
fig.savefig('cornbean_rotation_ratio.png')
```
