```python
import psycopg2
WEPPDB = psycopg2.connect(database='wepp', host='iemdb', user='nobody')
wcursor = WEPPDB.cursor()
IDEPDB = psycopg2.connect(database='idep', host='iemdb', user='nobody')
icursor = IDEPDB.cursor()

# Build cross reference of iatwps to huc12s
xref = {}

wcursor.execute("""SELECT model_twp, ST_X(ST_Centroid(the_geom)), ST_Y(ST_Centroid(the_geom)) from iatwp""")
for row in wcursor:
    icursor.execute("""SELECT huc_12 from ia_huc12 WHERE ST_Within(ST_SetSRID(ST_GeomFromText('POINT(%s %s)'),26915), geom)""",
                    (row[1], row[2]) )
    row2 = icursor.fetchone()
    xref[ row[0] ] = row2[0]
```

```python
wcursor = WEPPDB.cursor()
icursor = IDEPDB.cursor()
year = 2013
SCENARIO = 0

v1 = {}
wcursor.execute("""SELECT model_twp, sum(avg_loss) from results_by_twp where valid BETWEEN '%s-01-01' and '%s-12-31'
 GROUP by model_twp""" % (year, year))
for row in wcursor:
    v1[row[0]] = row[1]

v2 = {}
icursor.execute("""SELECT huc_12, sum(avg_loss) from results_by_huc12 where valid BETWEEN '%s-01-01' and '%s-12-31'
 and scenario = %s and avg_loss != 'NaN' GROUP by huc_12""" % (year, year, SCENARIO))
for row in icursor:
    v2[row[0]] = row[1]
```

```python
x = []
y = []
for twp in xref.keys():
    x.append( v1.get(twp,0) * 4.463)
    y.append( v2.get(xref[twp],0) * 4.463)
x = np.array(x)
y = np.array(y)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

(fig, ax) = plt.subplots(1,1)

(counts, yedges, xedges, res) = ax.hist2d(x, y, bins=40, cmin=1, cmax=50, cmap=cm.get_cmap('spectral'))
fig.colorbar(res)
ax.axhline(np.average(y), lw=2, color='yellow', zorder=2)
ax.axhline(np.average(y), lw=4, color='g', label='v2: %.1f' % (np.average(y),), zorder=1)
ax.axvline(np.average(x), lw=2, color='yellow', zorder=2)
ax.axvline(np.average(x), lw=4, color='g', label='v1: %.1f' % (np.average(x),), zorder=1)
ax.set_xlim(left=0)
ax.set_ylim(bottom=0)
ymax = ax.get_ylim()[1]
ax.grid()
ax.set_xlabel("IDEP Loss [tons/acre]")
ax.set_ylabel("IDEPv2 Loss [tons/acre]")
ax.set_title("IDEP v1 to v2 :: %s Average Soil Loss\nbased on prelim data 16 June 2014" % (year,))
ax.plot([0,ymax], [0,ymax], lw=2, color='r', label='1 to 1')
ax.legend()
fig.savefig("avgloss_%s.png" % (year,), dpi=300)
```
